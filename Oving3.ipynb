## Oppgave 3

Vi har funksjonen

$$
\theta = \cos^{-1}\left(\frac{c\tau}{d}\right)
$$

Sett

$$
f = \frac{c\tau}{d}
$$

Da kan vi skrive

$$
\theta = \cos^{-1}(f)
$$

---

### Feilforplantning

For små feil gjelder

$$
\Delta \theta \approx \frac{d\theta}{df}\,\Delta f
$$

Derivasjonen av $\cos^{-1}(f)$ er

$$
\frac{d\theta}{df} = -\frac{1}{\sqrt{1 - f^2}}
$$

Dermed får vi maksimal absolutt feil:

$$
|\Delta \theta|_{\text{maks}} =
\frac{|\Delta f|_{\text{maks}}}{\sqrt{1 - f^2}}
$$

---

### Sett inn feil i $f$

Fra tidligere:

$$
\left|\frac{\Delta f}{f}\right|_{\text{maks}} = 0.03
\quad \Rightarrow \quad
|\Delta f|_{\text{maks}} = 0.03 f
$$

Setter dette inn:

$$
|\Delta \theta|_{\text{maks}} =
\frac{0.03 f}{\sqrt{1 - f^2}}
$$

---

### Relativ feil i $\theta$

$$
\left|\frac{\Delta \theta}{\theta}\right|_{\text{maks}} =
\frac{0.03 f}{\theta \sqrt{1 - f^2}}
$$

---

### Skriv om med $\theta$

Siden

$$
f = \cos\theta
$$

får vi

$$
\sqrt{1 - f^2} = \sin\theta
$$

Setter inn:

$$
\left|\frac{\Delta \theta}{\theta}\right|_{\text{maks}} =
\frac{0.03 \cos\theta}{\theta \sin\theta}
$$

---

### Endelig uttrykk

$$
\boxed{
\left|\frac{\Delta \theta}{\theta}\right|_{\text{maks}} =
0.03 \cdot \frac{1}{\theta \tan\theta}
}
$$

```python
import numpy as np

nrand = int(1e5)

theta_values = np.array([5, 10, 20, 30, 45, 60, 80]) * np.pi/180

for theta0 in theta_values:
    
    # sann verdi
    f0 = np.cos(theta0)
    
    # uniform relativ feil i tau og d
    tauvec = 1 + 2*0.01*(np.random.rand(nrand) - 0.5)
    dvec   = 1 + 2*0.02*(np.random.rand(nrand) - 0.5)
    
    # c settes til 1, fordi c er nøyaktig
    fvec = f0 * tauvec / dvec
    
    # fjern verdier der f > 1, fordi arccos ikke er definert der
    fvec = fvec[fvec <= 1]
    
    thetavec = np.arccos(fvec)
    
    rel_error = np.abs((thetavec - theta0)/theta0)
    
    sim_maks = np.max(rel_error)
    
    teori = 0.03 / (theta0 * np.tan(theta0))
    
    print("theta =", theta0*180/np.pi, "grader")
    print("Simulert maks relativ feil =", sim_maks)
    print("Teoretisk relativ feil     =", teori)
    print()
```

## Oppgave 4

Vi har

$$
\theta = \cos^{-1}\left(\frac{c\tau}{d}\right)
$$

Sett

$$
f = \frac{c\tau}{d}
\quad \Rightarrow \quad
\theta = \cos^{-1}(f)
$$

---

### Steg 1: Feilforplantning

For standardavvik gjelder

$$
\sigma_\theta^2 = \left(\frac{d\theta}{df}\right)^2 \sigma_f^2
$$

Derivasjon:

$$
\frac{d\theta}{df} = -\frac{1}{\sqrt{1 - f^2}}
$$

Dermed

$$
\sigma_\theta = \frac{\sigma_f}{\sqrt{1 - f^2}}
$$

---

### Steg 2: Standardavvik i $f$

For

$$
f = \frac{c\tau}{d}
$$

og siden $c$ er nøyaktig:

$$
\left(\frac{\sigma_f}{f}\right)^2 =
\left(\frac{\sigma_\tau}{\tau}\right)^2 +
\left(\frac{\sigma_d}{d}\right)^2
$$

Setter inn:

$$
\left(\frac{\sigma_f}{f}\right)^2 =
(0.01)^2 + (0.02)^2 = 0.0005
$$

$$
\Rightarrow \frac{\sigma_f}{f} = \sqrt{0.0005} \approx 0.02236
$$

Altså

$$
\sigma_f = 0.02236\, f
$$

---

### Steg 3: Sett inn i $\sigma_\theta$

$$
\sigma_\theta = \frac{0.02236 f}{\sqrt{1 - f^2}}
$$

---

### Steg 4: Relativt standardavvik

$$
\frac{\sigma_\theta}{\theta} =
\frac{0.02236 f}{\theta \sqrt{1 - f^2}}
$$

Bruk

$$
f = \cos\theta, \quad \sqrt{1 - f^2} = \sin\theta
$$

Da får vi

$$
\frac{\sigma_\theta}{\theta} =
\frac{0.02236 \cos\theta}{\theta \sin\theta}
$$

---

### Endelig svar

$$
\boxed{
\frac{\sigma_\theta}{\theta}
=
0.02236 \cdot \frac{1}{\theta \tan\theta}
}
$$
