For å lage måleinstrumentet, har vi brukt motstander med resistansen R = 1 kΩ, med presisjonen 1%. Vi kan altså anta at resistansverdien har en uniform fordeling ±1%.

a. Grenseverdiene er [990, 1010], siden det er 1 prosent ned ogg opp fra 1kOhm. 

```python
std = 20/np.sqrt(12)
print(f"Standaravikegt er {std}")
```

```python
print(f"Relativt standarvik er {std/1000*100}%")
```

c. Resistansverdien blir summen av to uavhengige uniforme fordelte størrelser. Summen av to uniforme fordelinger gir en trekantfordeling. Den totale resistansen vil derfor være trekantfordelt på intervallet [990,1010] Ω, med høyest sannsynlighet rundt 1000 Ω og lavest sannsynlighet ved ytterpunktene.

```python
N = 1000000

R1 = np.random.uniform(495, 505, N)
R2 = np.random.uniform(495, 505, N)
R_tot = R1 + R2

middelverdi = np.mean(R_tot)
std = np.std(R_tot)
rel_std = (std / middelverdi) * 100

print(f"Middelverdi: {middelverdi:.2f} Ω")
print(f"Standardavvik: {std:.2f} Ω")
print(f"Relativt standardavvik: {rel_std:.3f} %")
```
